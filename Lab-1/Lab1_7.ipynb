1. What is the result of the following expression?

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

1) 0 * np.nan : ***результатом будет nan***, так как результат арифметических операций, один из операндов которых является nan, будет nan, согласно стандарту
2) np.nan == np.nan : результат будет Fasle, потому что nan (Not a Number) это такая переменная, которая вообще ничему не равна, в том числе не равна и самому nan

3) np.inf > np.nan : ***результатом будет Fasle***, потому что, согласно стандарту, любое сравнение с участием nan должно всегда возвращать False

4) np.nan - np.nan : ***результатом будет nan***, так как результат арифметических операций, один из операндов которых является nan, будет nan, согласно стандарту

5) np.nan in set([np.nan]) : ***в этому случае результатом будет True***, потому что мы проверяем идентичность двух объектов, а не сравниваем их с помощью ==. В итоге мы првоеряем на идентичность два одинаковых объекта и получаем в результате True

6) 0.3 == 3 * 0.1 : ***результатом будет False***. Это связано с представлением числе с плавающей точкой в памяти компьютера - у них теряется точность при вычислениях и возникает погрешность

````python
print(0 * np.nan)
print(np.nan == np.nan)
print(np.inf > np.nan)
print(np.nan - np.nan)
print(np.nan in set([np.nan]))
print(0.3 == 3 * 0.1)
````

8. Выведите минимальное и максимальное представимое значение для каждого скалярного типа numpy

````python
def print_min_max_scalar_values() -> None:
    print("Signed Integer Types")
    for int_type in [np.int_, np.intc, np.intp, np.int8, np.int16, np.int32, np.int64]:
        type_info = np.iinfo(int_type)
        print(f"{int_type.__name__:<5} : min is {type_info.min} , max is {type_info.max}")

    print("\nUnsigned Integer Types")
    for uint_type in [np.uint, np.uintp, np.uintc, np.uint8, np.uint16, np.uint32, np.uint64]:
        type_info = np.iinfo(uint_type)
        print(f"{uint_type.__name__:<6} : min is {type_info.min} , max is {type_info.max}")

    print("\nFloating Point Types")
    for float_type in [np.float_, np.float16, np.float32, np.float64, np.float128]:
        type_info = np.finfo(float_type)
        print(f"{float_type.__name__:<6} : min is {type_info.min} , max is {type_info.max}")


print_min_max_scalar_values()
````

9. Как отсортировать массив по n-му столбцу?

````python
def sort_by_column(array : np.ndarray, column : int, descending=False) -> np.ndarray:
    order_flag = 1 if not descending else -1
    return array[array[:, column].argsort()[::order_flag]]


random_matrix5x5 = np.random.randint(0, 10, (5, 5))

print(random_matrix5x5, '\n')

print(f"Sorted ASC\n{sort_by_column(random_matrix5x5, 0)}\n")
print(f"Sorted DESC\n{sort_by_column(random_matrix5x5, 0, descending=True)}")
````

10. Как определить, есть ли в данном двумерном массиве пустые столбцы?

````python
def has_null_columns(matrix : np.ndarray) -> bool:
    return ((~matrix.any(axis=0)).any())
'''
matrix.any(axis=0) - возвращает массив bool размер, которого количество столбцов в нашей матрице,
                     со значением True, где в соответствующем столбце есть хотя бы один не нулевой
                     элемент, и False в противоположном случае. Параметр axis=0 указывает на то,
                     что операция должна выполняться вдоль столбцов
~matrix.any(axis=0) - инвертирует значения в массиве, полученном на предыдущем шаге. Теперь True
                      означает, что столбец содержит все нули
(~matrix.any(axis=0)).any() - проверяет, есть ли хотя бы одно значение True в нашем массиве
'''


random_matrix4x5 = np.random.randint(0, 10, (4, 5))

matrix_with_null_columns = np.zeros_like(random_matrix4x5)
matrix_with_null_columns[::1, 1::2] = 1

print(f"{random_matrix4x5}\n\n{matrix_with_null_columns}\n")

print(f"Random matrix 4x5 has null columns - {has_null_columns(random_matrix4x5)}")
print(f"Matrix with null columns has null columns - {has_null_columns(matrix_with_null_columns)}")
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
def parse_blood_pressure(blood_pressure : str) -> pd.Series :
    systolic, diastolic = map(int, blood_pressure.split('/'))

    return pd.Series([systolic, diastolic])


df[["Systolic", "Diastolic"]] = df["Blood Pressure (s/d)"].apply(parse_blood_pressure)

stress_levels = [
                 "low",
                 "high"
                 ]

stress_levels_conditions = [
                            (df["Stress Levels"] <= 5),
                            (df["Stress Levels"] > 5)
                            ]

df["Stress Category"] = np.select(stress_levels_conditions, stress_levels)

stress_pressure_dependence = df.groupby("Stress Category").agg(
                                                               {
                                                                "Systolic" : ["mean"],
                                                                "Diastolic" : ["mean"]
                                                                }
                                                               )

stress_pressure_dependence.columns = ["Systolic", "Diastolic"]

low_systolic_mean = stress_pressure_dependence["Systolic"].loc[stress_pressure_dependence.index[1]]
low_diastolic_mean = stress_pressure_dependence["Diastolic"].loc[stress_pressure_dependence.index[1]]

high_systolic_mean = stress_pressure_dependence["Systolic"].loc[stress_pressure_dependence.index[0]]
high_diastolic_mean = stress_pressure_dependence["Diastolic"].loc[stress_pressure_dependence.index[0]]

print(stress_pressure_dependence, '\n')

ERROR_RATE = 5.

if abs(high_systolic_mean - low_systolic_mean) < ERROR_RATE and abs(high_diastolic_mean - low_diastolic_mean) < ERROR_RATE:
    print("Неправда")
else:
    print("Правда")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
BMI_30_percentile = df["BMI"].quantile(0.30)
BMI_60_percentile = df["BMI"].quantile(0.60)
BMI_75_percentile = df["BMI"].quantile(0.75)

df["Chronic Diseases"] = df["Chronic Diseases"].fillna("No")

group_from_30_to_60 = df[(df["BMI"] >= BMI_30_percentile) & (df["BMI"] <= BMI_60_percentile)]
group_more_75 = df[df["BMI"] > BMI_75_percentile]

amount_from_30_to_60 = group_from_30_to_60.shape[0]
amout_more_75 = group_more_75.shape[0]

from_30_to_65_with_chtonic_diseases = group_from_30_to_60[group_from_30_to_60["Chronic Diseases"] != "No"].shape[0]
more_75_with_chronic_diseases = group_more_75[group_more_75["Chronic Diseases"] != "No"].shape[0]

percent_from_30_to_60_with_chronic_diseases = 100 * from_30_to_65_with_chtonic_diseases / amount_from_30_to_60
percent_more_75_with_chronic_diseases = 100 * more_75_with_chronic_diseases / amout_more_75

ERROR_RATE = 1.5
in_the_margin_of_error = True if abs(percent_from_30_to_60_with_chronic_diseases - percent_more_75_with_chronic_diseases) <= ERROR_RATE \
                              else False

if percent_more_75_with_chronic_diseases > percent_from_30_to_60_with_chronic_diseases and not in_the_margin_of_error:
    print("Правда\n")
else:
    print("Неправда\n")

print("Процент встречаемости хронических заболеваний у людей с индексом массы тела")
print(f" - в пределах от 30 до 60 перцентиля - {percent_from_30_to_60_with_chronic_diseases:.2f}%")
print(f" - больше 75 перцентиля - {percent_more_75_with_chronic_diseases:.2f}%")
````
